# Metrics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Part 2: Put, get, delete operation latency

```python
dict = {'':['1', '2', '3', '4', '5'],
        'put':[451, 342, 396, 339, 544],
        'get':[221, 361, 205, 338, 396],
        'delete': [0, 0, 0, 0, 0],
       }

df = pd.DataFrame(dict)
  
display(df)
```

```python
df.describe()
```

```python
plt.plot([1,1,1,1,1], df["put"], label="put")
plt.plot([2,2,2,2,2], df["get"], label="get")
plt.plot([3,3,3,3,3], df["delete"], label="delete")
plt.xlabel("Trial #")
plt.ylabel("Latency time (measured in milliseconds)")
plt.title("File size 25MB")
plt.legend(loc="upper right")
fig = plt.figure(figsize =(10, 7))
 
# Creating plot
bp = df.boxplot(figsize=(7,5))
```

## Part 4: Wikipedia corpus

```python
dict2 = {'4': [18.16, 27.37, 24.09, 26.7, 18.6], '8': [19.89, 27.44, 31.2, 25.77, 30.04]}
df2 = pd.DataFrame(dict2)

plt.plot([1,1,1,1,1], [18.16, 27.37, 24.09, 26.7, 18.6], label="put")
plt.plot([2,2,2,2,2], [19.89, 27.44, 31.2, 25.77, 30.04], label="get")
plt.xlabel("Number of members")
plt.ylabel("Latency time (measured in milliseconds)")
plt.title("Wikipedia Corpus")
plt.legend(loc="upper right")
fig = plt.figure(figsize =(10, 7))
 
# Creating plot
bp = df2.boxplot(figsize=(7,5), showmeans=True)
bp.set_title('Wikipedia Corpus')
bp.set_xlabel('Number of members')
bp.set_ylabel('Latency time (measured in milliseconds)')

```
